```python
import os
from pathlib import Path
import statistics

# Diccionarios para almacenar los datos por grupo
group_data = {'consolidado': [], 'reciente': [], 'detectado': []}

# Rutas de las carpetas que contienen los archivos .txt
folders = {
    'consolidado': Path('datasetvih/VIHconsolidado'),
    'reciente': Path('datasetvih/VIHreciente'),
    'detectado': Path('datasetvih/VIHdetectado')
}

# Filtrar documentos por longitud mínima
min_length = 3000  # Ejemplo de umbral, ajustar según análisis previo

# Recorrer cada carpeta y procesar los archivos
for group, folder_path in folders.items():
    files = sorted(folder_path.glob('*.txt'))  # Obtener archivos .txt y ordenarlos

    # Recorrer todos los archivos .txt y calcular su longitud
    for index, file_path in enumerate(files):
        print(f"Archivo {file_path.name}")  # Imprimir el nombre del archivo
        try:
            with file_path.open('r', encoding='utf-8') as file:
                content = file.read()  # Leer el contenido del archivo
                length = len(content)  # Calcular la longitud del contenido
                
                if length >= min_length:  # Solo procesar si el documento supera el umbral de longitud
                    # Ejecutar el clasificador usando el path del archivo y el índice local
                    infected, score = clasificador.execute(str(folder_path), index)
                    print(f"Leyendo archivo: {file_path.name}")  # Imprimir el nombre del archivo

                    # Añadir la longitud, infected y score al grupo correspondiente
                    group_data[group].append((file_path.name, length, infected, score))
        except Exception as e:
            print(f"Error al procesar el archivo {file_path}: {e}")

# Función para calcular estadísticas de longitud de caracteres
def calculate_length_statistics(data):
    lengths_infected = [length for _, length, infected, _ in data if infected]
    lengths_non_infected = [length for _, length, infected, _ in data if not infected]

    if lengths_infected:
        avg_length_infected = sum(lengths_infected) / len(lengths_infected)
        stddev_length_infected = statistics.stdev(lengths_infected) if len(lengths_infected) > 1 else 0.0
    else:
        avg_length_infected = 0
        stddev_length_infected = 0

    if lengths_non_infected:
        avg_length_non_infected = sum(lengths_non_infected) / len(lengths_non_infected)
        stddev_length_non_infected = statistics.stdev(lengths_non_infected) if len(lengths_non_infected) > 1 else 0.0
    else:
        avg_length_non_infected = 0
        stddev_length_non_infected = 0

    return avg_length_infected, stddev_length_infected, avg_length_non_infected, stddev_length_non_infected

# Imprimir los datos de cada grupo
for group, data in group_data.items():
    if data:  # Verificar si hay datos en el grupo
        total_infected = sum(1 for _, _, infected, _ in data if infected)
        total_documents = len(data)
        average_score = sum(score for _, _, _, score in data) / total_documents
        average_length = sum(length for _, length, _, _ in data) / total_documents
        avg_length_infected, stddev_length_infected, avg_length_non_infected, stddev_length_non_infected = calculate_length_statistics(data)
        
        print(f"Documentos VIH {group.capitalize()} - {total_infected} correctos de {total_documents} documentos, Media de score: {average_score:.2f}, Media de caracteres: {average_length:.2f}:")
        print(f"  Media de caracteres (Correctos): {avg_length_infected:.2f}, Desviación típica: {stddev_length_infected:.2f}")
        print(f"  Media de caracteres (Incorrectos): {avg_length_non_infected:.2f}, Desviación típica: {stddev_length_non_infected:.2f}")
        for file_name, length, infected, score in data:
            status = "Correcto" if infected else "Incorrecto"
            print(f"{file_name}: {length} caracteres - {status}, Score: {score:.2f}")
        print()  # Añadir una línea en blanco para separar los grupos
```
